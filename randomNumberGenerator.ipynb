# Perfect Random File

```python
randomList = []
counter = 0

while counter < 1000:
    rand = random.uniform(0,2)
    temp = str(counter)+ ", "+str(rand)+", "+str(rand)
#     temp.join(", ")
  
#     temp = [str(counter", ",str(rand)", "str(rand))]
#     counter += 1
    randomList.append(temp)
    print (temp)
    counter+=1
```

```python
# places = ['Berlin', 'Cape Town', 'Sydney', 'Moscow']

with open('randomPerfect.txt', 'w') as filehandle:
    for listitem in randomList:
        filehandle.write('%s\n' % listitem)
```
